```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
```

```python
def plotSin():
    xValues = np.arange(-15.,15.,0.01)
    yValues = np.sin(xValues)
    fig, ax = plt.subplots()
    plt.plot(xValues,yValues)
    plt.xlabel('x')
    plt.ylabel('sin(x)')
    plt.show()
```

```python
plotSin()
```

```python
def plotU():
    rValues = np.linspace(0.6, 10, 1000)
    uValues = (4*12)*(rValues**(12))**(-1)-2*(4**6)*(rValues**1)**(-1)
    fig, ax = plt.subplots()
    ax.set_xticks([])
    ax.set_yticks([])
    plt.plot(rValues,uValues)
    plt.xlabel('R')
    plt.ylabel('U(R)')
    plt.show()
```

```python
plotU()
```

```python
def func(x, b, c):
    R=8.314 
    #[R]=(J)/(mol*K)
    #[a]=(cm^3)*(mol^2)/(J)
    #[b]=(cm^3)/(mol)
    #[t]=(K)
    return b*x+c #b-a/(R*t)
```

```python
def curveFitB():
    dataB = (17550, 27672, 8136.7)#100, 200, 300, 400, 500, 600)
    dataT = (12, 0, 36.2)#-160, -35, -4.2, 9.0, 16.9, 21.3)
    valueT = np.linspace(0, 50, 50)
    
    popt, pcov = curve_fit(func, dataT, dataB)
    print(popt)
    
    plt.figure()
    plt.plot(valueT, func(valueT, *popt), 'b-', 
             label='fit: b=%5.3f, c=%5.3f' % tuple(popt))
    plt.plot(dataT, dataB, 'r+', label='original data')
    
    plt.xlabel('Temperature [K]')
    plt.ylabel('B(T) [cm^3/mol]')
    plt.legend()
    plt.show()
```

```python
curveFitB()
```

```python
def curveFitBZoomed():
    dataT = (100, 200, 300, 400, 500, 600)
    dataB = (-160, -35, -4.2, 9.0, 16.9, 21.3)
    valueT = np.linspace(90, 900, 1000)
    
    popt, pcov = curve_fit(func, dataT, dataB)
    print(popt)
    
    plt.figure()
    plt.plot(valueT, func(valueT, *popt), 'b-', 
             label='fit: a=%5.3f, b=%5.3f' % tuple(popt))
    plt.plot(dataT, dataB, 'r+', label='original data')
    
    plt.xlabel('Temperature [K]')
    plt.ylabel('B(T) [cm^3/mol]')
    plt.legend()
    plt.show()
```
